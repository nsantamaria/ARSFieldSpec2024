```python
def divide_dataframe(df, label):
    # Ensure the first column is named "Wavelength"; adjust if your dataframe is different
    wavelength_column = df.columns[0]
    
    # Initialize lists with the first column to include in both DataFrames
    top_leaf_columns = [wavelength_column]
    middle_leaf_columns = [wavelength_column]
    
    # Iterate over the columns starting from the second one
    for col in df.columns[1:]:
        if "TL" in str(df.iloc[0][col]):
            top_leaf_columns.append(col)
        elif "ML" in str(df.iloc[0][col]):
            middle_leaf_columns.append(col)
    
    # Create the new DataFrames based on the columns lists
    top_leaf_df = df[top_leaf_columns].copy()
    middle_leaf_df = df[middle_leaf_columns].copy()
    
    # Optionally, you can modify DataFrame names here using the label
    top_leaf_df.name = f"Top_Leaf_{label}"
    middle_leaf_df.name = f"Middle_Leaf_{label}"
    #rename first column to "Wavelength" for top_leaf_df and middle_leaf_df
    
    #Drop first row from top_leaf_df and middle_leaf_df
    middle_leaf_df = middle_leaf_df.drop(df.index[0]).reset_index(drop=True)
    top_leaf_df = top_leaf_df.drop(df.index[0]).reset_index(drop=True)
    
    
    return top_leaf_df, middle_leaf_df
    

# Initialize dictionaries to store the divided DataFrames
top_leaf = {}
middle_leaf = {}

# Iterate through the original dictionary, apply the function, and store the results
for label, df in data_2023.items():
    top_leaf_df, middle_leaf_df = divide_dataframe(df, label)
    top_leaf[label] = top_leaf_df
    middle_leaf[label] = middle_leaf_df

# Assuming 'top_leaf' and 'middle_leaf' are your existing dictionaries

# Initialize an empty dictionary to hold the unified data with new key descriptors
unified_data = {}
#convert top and middle leaf to numeric
for key in top_leaf:
    top_leaf[key] = top_leaf[key].apply(pd.to_numeric, errors='coerce')
    middle_leaf[key] = middle_leaf[key].apply(pd.to_numeric, errors='coerce')

# Update the keys and merge the dictionaries
for key in top_leaf:
    # Add the "Top Leaf" descriptor to the key for the top leaf data
    new_key_top = f"{key}, Top Leaf"
    unified_data[new_key_top] = top_leaf[key]
    

for key in middle_leaf:
    # Add the "Middle Leaf" descriptor to the key for the middle leaf data
    new_key_middle = f"{key}, Middle Leaf"
    unified_data[new_key_middle] = middle_leaf[key]
    
```

```python
import os
import matplotlib.pyplot as plt
import pandas as pd

# Base directory where plots will be saved
base_dir = 'Mean_Intensity_Plots'

# Ensure the DataFrame is in the correct numeric format
def make_mean_intensity_plots(days, plot_groups):
    # Create the base directory if it doesn't exist
    if not os.path.exists(base_dir):
        os.makedirs(base_dir)

    for day_name, df in days.items():
        #Convert all columns to numeric
        df = df.apply(pd.to_numeric, errors='ignore')
    
        day_dir = os.path.join(base_dir, day_name)
        if not os.path.exists(day_dir):
            os.makedirs(day_dir)
        print(day_name)
        
        # Now working with columns excluding 'Wavelength'
        for plot_id, columns in plot_groups.items():
            # Here, ensure 'columns' only contains columns after 'Wavelength'
            # Adjust 'columns' list to exclude any that are not in 'df.columns'
            columns = [col for col in columns if col in df.columns[1:]]  # Skip 'Wavelength'
            
            if not columns:  # If 'columns' is empty after filtering, skip to the next iteration
                continue
            

            # Now 'columns' is ready for use, and does not include 'Wavelength'
            mean_values = df[columns].mean(axis=1)
            

            # Create a new figure for each plot ID
            plt.figure(figsize=(10, 6))

            # Plot the mean values. If your DataFrame includes a 'Wavelength' column, use it as the x-axis.
            if 'Wavelength' in df.columns:
                plt.plot(df['Wavelength'], mean_values, label=f'Mean Intensity')
                plt.xlabel('Wavelength (nm)')
            else:
                plt.plot(mean_values, label=f'Plot ID {plot_id}')
                plt.xlabel('Sample Index')
            
            plt.ylabel('Mean Intensity Value')
            plt.title(f'Mean Intensity for Plot ID {plot_id}: {day_name}')
            plt.legend()

            # Save the plot to the appropriate directory instead of showing it
            plot_filename = f'plot_id_{plot_id}.png'
            plt.savefig(os.path.join(day_dir, plot_filename))

            # Clear the current figure after saving it to avoid overlap with next plots
            plt.clf()

# Example usage
make_mean_intensity_plots(days, plot_groups)

```

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.linear_model import LinearRegression

base_dir = 'VI_correlations'

def sanitize_filename(name):
    """
    Sanitize a string to be used as a filename by replacing or removing characters
    that are not valid or advisable in filenames.
    """
    # Replace or remove invalid or problematic characters
    name = name.replace('%', 'perc')  # Replace '%' with 'perc'
    name = name.replace(',', '')      # Remove commas
    name = name.replace(' ', '_')     # Replace spaces with underscores
    name = name.replace('/', '_or_')  # Replace slashes with '_or_' to avoid directory issues
    return name.lower()  # Return the sanitized name in lowercase




def get_treatment_from_column(column_name):
    plot_number_match = re.search(r"N(\d{3})\dP", column_name)
    if plot_number_match:
        plot_number = plot_number_match.group(1)
        return plot_treatment_map.get(plot_number, 'Unknown')

def aggregate_and_plot(mean_vi_values_by_day, harvest, root_means):
    # Ensure the base directory exists
    if not os.path.exists(base_dir):
        os.makedirs(base_dir)

    for day_name, vi_values in mean_vi_values_by_day.items():
        day_dir = os.path.join(base_dir, day_name)
        if not os.path.exists(day_dir):
            os.makedirs(day_dir)
        
        def plot_relationship(X, y, vi_name, y_label):
            if not X or not y:
                print(f"No data available for {vi_name} vs {y_label} for {day_name}.")
                return
            
            X = np.array(X).reshape(-1, 1)
            y = np.array(y)
            model = LinearRegression().fit(X, y)
            predictions = model.predict(X)
            rmse = np.sqrt(mean_squared_error(y, predictions))
            r2 = r2_score(y, predictions)
            
            plt.figure(figsize=(10, 6))
            plt.scatter(X, y, color='blue', label='Actual')
            plt.plot(X, predictions, color='red', label='Predicted')
            plt.title(f'{vi_name} vs {y_label}\n{day_name}\nRMSE: {rmse:.2f}, R: {r2:.2f}')
            plt.xlabel(vi_name)
            plt.ylabel(y_label)
            plt.legend()
            # Define plot filename
            plot_filename = f"{sanitize_filename(vi_name)}_{sanitize_filename(y_label)}.png"
            plt.savefig(os.path.join(day_dir, plot_filename))
            plt.close()

        all_vi_names = set().union(*(d.keys() for d in vi_values.values()))

        for vi_name in all_vi_names:
            X, y_lists = {}, {}
            for plot_number, vis in vi_values.items():
                vi_value = vis.get(vi_name)
                if vi_value is not None:
                    for y_label, df, key in [
                        ('Grain Yield, 0% MC kg/ha', harvest, 'Grain Yield, 0% MC kg/ha'),
                        ('Grain Yield, 15.5% MC kg/ha', harvest, 'Grain Yield, 15.5% MC kg/ha'),
                        ('Annual ET, mm', harvest, 'Annual ET, mm'),
                        ('WP (kg/ha/mm)0MC', harvest, 'WP (kg/ha/mm)0MC'),
                        ('Mean Sum of Root Length (cm)', root_means, 'Sum of Length(cm)'),
                        ('Root Mean Mass (g)', root_means, 'mass_g')]:
                        row = df[df['Group'].astype(str) == str(plot_number)]
                        if not row.empty and key in row:
                            y_lists.setdefault(y_label, []).append(row[key].values[0])
                            X.setdefault(y_label, []).append(vi_value)
            
            for y_label, ys in y_lists.items():
                plot_relationship(X[y_label], ys, vi_name, y_label)

# Assuming 'mean_vi_values_by_day', 'harvest', and 'root_means' are defined
aggregate_and_plot(vis_by_day, harvest, root_means)
```
