```python
def modify_dataframe_include_leaf_position(df):
    # Assuming the first row contains leaf position indicators (TL, ML, etc.)
    leaf_position_row = df.iloc[0]
    # Create a new column 'Leaf_Position' to store this information
    df = df.assign(Leaf_Position=pd.NA)
    for col in df.columns:
        if "TL" in str(leaf_position_row[col]):
            df['Leaf_Position'] = 'Top Leaf'
        elif "ML" in str(leaf_position_row[col]):
            df['Leaf_Position'] = 'Middle Leaf'
    # Drop the first row and reset the index
    df = df.drop(df.index[0]).reset_index(drop=True)
    return df
```

```python
import pandas as pd
import xarray as xr

def dataframe_to_xarray_with_leaf_position(df, day_label):
    # Extract leaf position indicators from the first row and create a mapping for column names
    leaf_position_map = {col: df.iloc[0][col] for col in df.columns[1:]}  # Skip 'Wavelength'
    
    # Drop the first row (leaf position indicators) and reset the index
    df = df.drop(index=0).reset_index(drop=True)
    
    # Rename columns to include leaf position, except for 'Wavelength'
    new_columns = {'Wavelength': 'Wavelength'}
    for col, position in leaf_position_map.items():
        new_columns[col] = f"{col}_{position}"  # Example: 'N111100001.asd_TL'
    df = df.rename(columns=new_columns)
    
    # Convert the modified DataFrame to an Xarray Dataset, setting 'Wavelength' as a non-dimension coordinate
    ds = xr.Dataset.from_dataframe(df.set_index('Wavelength'))
    
    # Add day label as a coordinate for easy access
    ds = ds.assign_coords(day=day_label)
    return ds

# Assuming your DataFrame (df) is loaded and processed
datasets = []
day_labels = ['Day1', 'Day2', 'Day3']  # Example day labels
dataframes = [df_day1, df_day2, df_day3]  # Your DataFrames here

for df, label in zip(dataframes, day_labels):
    ds = dataframe_to_xarray_with_leaf_position(df, label)
    datasets.append(ds)

# Combine all datasets along a new dimension ('time') if applicable
combined_ds = xr.concat(datasets, dim='time')
```
